```python
#loading the diabetes csv file
path_1 = '/content/gdrive/My Drive/diabetes.csv'
```

```python
import keras
import pandas
from keras.models import Sequential
from keras.layers.core import Dense, Activation

# loading dataset
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np

dst = pd.read_csv(path_1, header=None).values

A_train, A_test, B_train, B_test = train_test_split(dst[:,0:8], dst[:,8],
                                                    test_size=0.25, random_state=87)
np.random.seed(155)
a = Sequential() # create model
a.add(Dense(20, input_dim=8, activation='relu')) # hidden layer
a.add(Dense(1, activation='sigmoid')) # output layer
a.compile(loss='binary_crossentropy', optimizer='adam', metrics=['acc'])
a_fitted = a.fit(A_train, B_train, epochs=100,
                                     initial_epoch=0)
print(a.summary())
print(a.evaluate(A_test, B_test))
```

```python
b = Sequential() # create model
b.add(Dense(20, input_dim=8, activation='relu')) # hidden layer
b.add(Dense(10, input_dim=8, activation='relu'))
b.add(Dense(5, input_dim=8, activation='relu'))
b.add(Dense(1, activation='sigmoid')) # output layer
b.compile(loss='binary_crossentropy', optimizer='adam', metrics=['acc'])
b_fitted = b.fit(A_train, B_train, epochs=100,
                                     initial_epoch=0)
print(b.summary())
print(b.evaluate(A_test, B_test))
```
