```python
import keras
import pandas
from keras.models import Sequential
from keras.layers.core import Dense, Activation


# loading dataset
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np

dst = pd.read_csv(path_2)

#labeling the output data from string to numerical value
from sklearn.preprocessing import LabelEncoder

lb_m = LabelEncoder()
dst["diagnosis_code"] = lb_m.fit_transform(dst["diagnosis"])
dst["diagnosis_code"].value_counts()

#Column  'id' is specific to each entry, it is not included in the input.
A = dst[['radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']]

b = dst['diagnosis_code']
```

```python
#Split the data
A_train, A_test, b_train, b_test = train_test_split(A, b, stratify=b, test_size=0.2,shuffle=True, random_state=3)

#training the model
np.random.seed(155)
 # creating a model
a = Sequential()
a.add(Dense(30, input_dim=30, activation='relu')) # hidden layer
a.add(Dense(1, activation='sigmoid')) # output layer
a.compile(loss='binary_crossentropy', optimizer='adam', metrics=['acc'])
a_fit = a.fit(A_train, b_train, epochs=100,
                                     initial_epoch=0)
print(a.summary())
print(a.evaluate(A_test, b_test))
```

```python
#normalize the data using StandardScaler
from sklearn.preprocessing import StandardScaler
scal = StandardScaler()
scal.fit(A)
A_scaled = scal.transform(A)
```

```python
A_train, A_test, b_train, b_test = train_test_split(A_scaled, b, stratify=b, test_size=0.2,shuffle=True, random_state=5)

#training the model with normalize inpt
np.random.seed(155)
a = Sequential() # create model
a.add(Dense(30, input_dim=30, activation='relu')) # hidden layer
a.add(Dense(1, activation='sigmoid')) # output layer
a.compile(loss='binary_crossentropy', optimizer='adam', metrics=['acc'])
a.fitted = a.fit(A_train, b_train, epochs=100,
                                     initial_epoch=0)
print(a.summary())
print(a.evaluate(A_test, b_test))
```
